```python
%matplotlib inline
import os
import numpy as np
import pandas as pd
from scipy import stats, integrate
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)



HOME = os.path.expanduser("~")
DATA = HOME + "/data_buffer/WebQA/"
QA = DATA + "QA_dataset.tsv"
HEALTH = DATA + "Health_dataset.tsv"
#paths = [DATA+"corpus_0_train.tsv", DATA+"corpus_0_test.tsv",
       # DATA+"corpus_0_dev.tsv"]
```

```python
QA_lines = open(QA).read().splitlines()
HEALTH_lines = open(HEALTH).read().splitlines()

QA_lengths = []
for line in QA_lines:
    question = line.split('\t')[0].split()
    answer = line.split('\t')[1].split()
    QA_lengths.append([len(question), len(answer)])
    
HEALTH_lengths = []
for line in HEALTH_lines:
    question = line.split('\t')[0].split()
    answer = line.split('\t')[1].split()
    HEALTH_lengths.append([len(question), len(answer)])
```

```python
QA_answer_lengths = []
QA_question_lengths = []

HEALTH_answer_lengths = []
HEALTH_question_lengths = []

for data in QA_lengths:
    QA_question_lengths.append(data[0])
    QA_answer_lengths.append(data[1])

    
for data in HEALTH_lengths:
    HEALTH_question_lengths.append(data[0])
    HEALTH_answer_lengths.append(data[1])
    

total_question_lengths = [data[0] for data in HEALTH_lengths] + [data[0] for data in QA_lengths]
total_answer_lengths = [data[1] for data in HEALTH_lengths] + [data[1] for data in QA_lengths]
    
total_combined_lengths = [sum(data) for data in HEALTH_lengths] + [sum(data) for data in QA_lengths]
QA_combined_lengths = [sum(data) for data in QA_lengths]
HEALTH_combined_lengths = [sum(data) for data in HEALTH_lengths]
```
