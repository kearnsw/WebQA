```python
import pickle as pkl
from WebQA.core.QA_Page import QA_Page
from WebQA.pages.Medhelp import MedhelpPage
from WebQA.pages.Healthtap import HealthtapPage
from WebQA.core.Post import Post, Question, Answer
from WebQA.core.User import User
import pandas as pd
import sys
```

```python
with open("../data/qa.pkl", "rb") as f:
    pages = pd.read_pickle(f)
```

```python
questions = []
answers = []
for page in pages:
    if page and page.question != "":
        if len(page.question.text) < 100 and page.answers:
            medical_answers = [answer for answer in page.answers]
            if medical_answers:
                questions.append(page.question.text.strip())
                answers.append(medical_answers[0].text.strip())
            else:
                continue
                #answers.append(page.answers[0].text.strip())
```

```python
no_answer_count = 0
for page in pages:
    if page and not page.answers:
        no_answer_count += 1
print(no_answer_count)
print(len(pages) - no_answer_count)
```

```python
data = pd.DataFrame({"Question": questions, "Answer": answers})
```

```python
pd.set_option('display.max_colwidth', -1)
print(len(data))
data
```
