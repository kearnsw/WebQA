```python
import pickle as pkl
from medhelp.core.pages import QA_Page
import medhelp.core.posts
from medhelp.core import posts
from medhelp.core import user
import pandas as pd
import sys

# For backward compatibility of older data backups
sys.modules['posts'] = posts
sys.modules['user'] = user
```

```python
with open("../data/qa.pkl", "rb") as f:
    pages = pd.read_pickle(f)
```

```python
questions = []
answers = []
for page in pages:
    if page:
        if len(page.question.text) < 400 and page.answers:
            medical_answers = [answer for answer in page.answers if answer.user.isMedical]
            if medical_answers:
                questions.append(page.question.text.strip())
                answers.append(medical_answers[0].text.strip())
            else:
                continue
                #answers.append(page.answers[0].text.strip())
```

```python
no_answer_count = 0
for page in pages:
    if page and not page.answers:
        no_answer_count += 1
print(no_answer_count)
print(len(pages) - no_answer_count)
```

```python
data = pd.DataFrame({"Question": questions, "Answer": answers})
```
